```python
# Get the total number of rows (calls) in the DataFrame 'df'.
total_calls = df.shape[0]

# Group the DataFrame 'df' by the 'zip' column and calculate the size (number of calls) for each group.
calls_by_zip = df.groupby('zip').size()

# Calculate the percentage of calls for each zip code by dividing the number of calls in each zip code by the total number of calls.
percentage_by_zip = (calls_by_zip / total_calls) * 100

# Define a threshold percentage to determine which zip codes are considered relevant.
threshold = 0.1

# Filter the 'percentage_by_zip' Series to include only zip codes with a percentage greater than or equal to the threshold.
relevant_zips = percentage_by_zip[percentage_by_zip >= threshold].index

# Update the DataFrame 'df' to include only rows where the 'zip' column value is in the list of relevant zip codes.
df = df[df['zip'].isin(relevant_zips)]
```

## üí° Feature Engineering üí°

In this section, we will focus on the art and science of feature engineering. Feature engineering is a crucial step in the machine learning pipeline, and it often requires domain knowledge to create the most relevant features. By transforming, binning, and creating interactions among features, we can improve the model's performance and interpretability.

### New Features to Design:

- **Title Splitting**: The `title` column will be divided into two new columns:
  - `cat_1`: This will represent the primary category of the call.
  - `cat_2`: This will represent the sub-category of the call.

- **Timestamp Decomposition**: The `timestamp` column will be split into four distinct columns to capture various temporal aspects:
  - `monthDay`: Specific day when the call was made.
  - `year`: Year of the call.
  - `month`: Month when the call was recorded.
  - `hour`: The specific hour of the day when the call was made.
  - `weekDay`: Day of the week when the call was recorded.

Let's dive in and start engineering these features!

```python
# Splitting the strings in the 'title' column into two parts based on the first occurrence of ': '.
# This will create two separate lists, one for 'category_1' and another for 'category_2'.
category_1, category_2 = zip(*[title.split(': ', 1) for title in df.title])

# Create two new columns 'cat_1' and 'cat_2' in the DataFrame 'df' and assign the values from the respective lists.
df['cat_1'] = category_1  # Assigns the first part of the split to 'cat_1'.
df['cat_2'] = category_2  # Assigns the second part of the split to 'cat_2'.
```

```python
# Convert the 'timeStamp' column in the DataFrame 'df' to datetime format.
df['timeStamp'] = pd.to_datetime(df['timeStamp'])

# Extract various components of the timestamp and create new columns for each.
df['monthDay'] = df['timeStamp'].dt.day     # Extract the day of the month.
df['year'] = df['timeStamp'].dt.year         # Extract the year.
df['month'] = df['timeStamp'].dt.month       # Extract the month.
df['hour'] = df['timeStamp'].dt.hour         # Extract the hour of the day.
df['weekDay'] = df['timeStamp'].dt.dayofweek # Extract the day of the week as an integer (0 = Monday, 6 = Sunday).
```

## üß™ Hypothesis Testing: Analyzing Anomalies in 911 Call Categories üìû

In our data-driven quest to uncover insights, we now delve into a pivotal analytical stage: Hypothesis Testing. Our goal? To unearth any anomalous behavior in the distribution of 911 call categories across different localities.

### üéØ Hypothesis Objective:

We aim to ascertain if certain localities or zip codes demonstrate an unusual proportion of specific call categories. A deviation from the norm could hint at unique regional dynamics, infrastructure issues, or varying public perceptions which might be pivotal for administrative or strategic decisions.

### üìà Approach and Methodology:

1. **Grouping by Zip Code**: Our primary unit of analysis is the zip code, representing distinct localities. By grouping our data accordingly, we're able to dissect the volume of calls emanating from each region.

2. **Quantifying Call Categories**: For every zip code, we have tallied the volume of calls pertaining to each category. This gives us a clear, numerical insight into the distribution and dominance of call types in every region.

3. **Statistical Foundations**:
   - **Mean Calculation**: For each call category, we compute the average proportion across all zip codes. This serves as our benchmark ‚Äì the 'expected' behavior.
   - **Standard Deviation**: To understand the variability and spread of our data, we calculate the standard deviation for each call category. This metric offers insights into how much individual zip code proportions deviate from our established mean.
   
4. **Identifying Anomalies**:
   - With our mean and standard deviation at hand, we establish a threshold: any zip code whose call category proportion exceeds two standard deviations from the mean is flagged as anomalous. This boundary is based on the empirical rule, ensuring we capture significant deviations.

### üåê Implications:

Identifying such anomalies can be instrumental for local authorities. Whether it's deducing the need for enhanced infrastructure, recognizing unique regional challenges, or tailoring public awareness campaigns ‚Äì understanding these anomalies translates to more informed and effective decision-making. As we proceed, we'll visualise and interpret these findings to glean actionable insights.

```python
# Group the DataFrame 'df' by both 'zip' and 'cat_1', then calculate the count of each category within each zip code.
cat_zip_counts = df.groupby(['zip', 'cat_1']).size().reset_index(name='count_per_cat')

# Group the DataFrame 'df' by 'zip' and calculate the total number of calls for each zip code.
zip_counts = df.groupby('zip').size().reset_index(name='total_calls_per_zip')

# Merge the two DataFrames ('cat_zip_counts' and 'zip_counts') on the 'zip' column to combine the count per category and total calls per zip.
merged = pd.merge(cat_zip_counts, zip_counts, on='zip')

# Calculate the proportion of calls for each category within each zip code by dividing 'count_per_cat' by 'total_calls_per_zip'.
merged['proportion'] = merged['count_per_cat'] / merged['total_calls_per_zip']

# Calculate the mean proportions for each call category ('cat_1') by grouping the 'merged' DataFrame and taking the mean of the 'proportion' column.
mean_proportions = merged.groupby('cat_1')['proportion'].mean()

# Calculate the standard deviations of proportions for each call category ('cat_1') by grouping the 'merged' DataFrame and taking the standard deviation of the 'proportion' column.
std_proportions = merged.groupby('cat_1')['proportion'].std()
```

```python
# Define a function 'is_anomaly' that takes a row as input and checks if the row is an anomaly based on mean and standard deviation.
def is_anomaly(row):
    # Get the mean proportion for the category 'cat_1' from the 'mean_proportions' dictionary.
    mean = mean_proportions[row['cat_1']]
    
    # Get the standard deviation for the category 'cat_1' from the 'std_proportions' dictionary.
    std = std_proportions[row['cat_1']]
    
    # Check if the absolute difference between the row's proportion and the mean is greater than 2 times the standard deviation.
    return abs(row['proportion'] - mean) > 2 * std

# Apply the 'is_anomaly' function to each row of the 'merged' DataFrame, and create a new 'anomaly' column to store the result.
merged['anomaly'] = merged.apply(is_anomaly, axis=1)
```

## üìä Data Visualization: Unveiling the Anomalies in 911 Call Categories üö®

Upon calculating anomalies, visual techniques play a pivotal role in discerning patterns and variances that might otherwise go unnoticed. In our analysis, a box plot serves as a foundational tool for this visualization. 

The box plot provides a graphical view of the distribution of the data, its central value, and its spread. In the context of our data, it allows us to compare the distribution of call categories' proportions for each zip code against the overall distribution, highlighting outliers and potential areas of concern.

```python
# Create a new figure for the plot with a specified size.
plt.figure(figsize=(12, 8))

# Create a box plot to visualize the distribution of proportions by call category ('cat_1').
sns.boxplot(x='cat_1', y='proportion', data=merged)

# Create a swarm plot to highlight anomalies in red. It displays individual data points for anomalies.
sns.swarmplot(x='cat_1', y='proportion', data=merged[merged['anomaly']], color='red', size=8)

# Set the title, ylabel, and xlabel for the plot.
plt.title("Distribution of Proportions by Call Category")
plt.ylabel("Call Proportions")
plt.xlabel("Call Category")

# Show the plot.
plt.show()
```

### üìú Tabular Insights:

After our visual deep dive, a detailed table assists in providing a granular breakdown of our findings:

Here's a brief rundown on interpreting the table:
- **Zip Code**: Represents distinct localities or regions.
- **Category**: Specifies the type or nature of the 911 call.
- **Proportion of Calls in Category**: This column shows the proportion of a particular category of calls in a specific zip code.
- **Average Proportion Across All Zip Codes**: This gives the average proportion of a specific category of calls across all zip codes.

For ease of interpretation, the background color of the cells in the "Proportion of Calls in Category" column changes based on its value:
- üî¥ **Red**: The proportion in that zip code is higher than the average proportion across all zip codes.
- üü¢ **Green**: The proportion in that zip code is lower than the average proportion.

This color-coding simplifies the process of identifying anomalies. A significant number of red or green cells for a specific zip code might indicate that the region has a unique pattern of 911 calls.

```python
merged = merged.merge(mean_proportions,how='left',on='cat_1')
merged = merged.merge(std_proportions,how='left',on='cat_1')
merged.rename(columns={'proportion_x':'proportion','proportion_y':'mean_proportion','proportion':'std_proportion'},inplace=True)
results = merged[merged.anomaly==1].groupby('zip')[['cat_1','proportion','mean_proportion']].apply(lambda x: x)
```

```python
def highlight_cell(x):
    c1 = 'proportion'
    c2 = 'mean_proportion'
    color = 'white'  # default color
    if x[c1] > x[c2]:
        color = 'red'
    elif x[c1] < x[c2]:
        color = 'green'
    return [f'background-color: {color}' if col == c1 else '' for col in x.index]

styled_df = results.style.apply(highlight_cell, axis=1)
styled_df
```
